Reshape data to index date column and drop duplicates

```
# Load our CSV data
import io
data = pd.read_csv(io.BytesIO(uploaded['EURUSD1h1y.csv']))
data.columns = ['Date','open','high','low','close','AskVol']
data.Date = pd.to_datetime(data.Date, format='%d.%m.%Y %H:%M:%S.%f')
data = data.set_index(pd.to_datetime(data.Date))
prices = data.drop_duplicates(keep=False)
data 
```

Bollingerbands

```
from talib import BBANDS
bbup, bbmid, bblow = BBANDS(data.close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
```

EMA

```
from talib import EMA
ema = EMA(data.close, timeperiod=30)
emadf = pd.DataFrame(ema).reset_index()
emadf = emadf.drop(columns=['Date'])
emadf.columns = ["ema"]
frame=[data,emadf]
data = pd.concat(frame)
data
```
