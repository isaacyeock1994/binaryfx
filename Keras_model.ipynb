Read data in a pandas data frame

```
import io
import pandas as pd
import numpy as np
df = pd.read_csv(io.BytesIO(uploaded['EURUSD15m1y41c.csv']))
```

```
df = df.drop(columns=['Date'])
```

Split data and get the shape 

```
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
#df = df.to_numpy()
X = df[:,0:40]

scaler = MinMaxScaler()
a = scaler.fit(X)
X = scaler.transform(X)


#scale X
Y = df[:,41]
X_train, X_test, y_train, y_test = train_test_split(df, Y, test_size=0.2)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```
import tensorflow as tf
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(42, input_dim=42, activation='relu'),
  tf.keras.layers.Dense(8, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['binary_accuracy'])
```

Train and evaluate the model

```
model.fit(X_train, y_train, epochs=100)
model.evaluate(X_test,  y_test, verbose=2)
```

Test the model

```
y_pred = model.predict_classes(X_test)
y_pred
```

Evaluate model using the confusion matrix

```
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test.round(),y_pred.round())
print(cm)
```
